```python
from sklearn.model_selection import KFold, GridSearchCV
from sklearn.linear_model import Lasso, Ridge
from sklearn.pipeline import make_pipeline
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import PolynomialFeatures
import numpy as np
from math import sqrt
```

```python
# Начальные данные 
x1_coords = [11,11,11,11,7,7,7,7,12.4,5.6,9,9,9,9,9,9,9,9,9,9,11]
x2_coords = [30,30,20,20,30,30,20,20,25,25,33.4,16.6,25,25,25,25,25,25,25,25,30]
u_list = [0.001886773,0.001888856,0.001839001,0.001840592,0.001861987,0.001858253,0.001799142,0.001809188,0.001869552,0.001828953,0.001876969,0.001792872,0.001852198,0.001849231,0.001860251,0.00184314,0.001878281,0.00185562,0.001851013,0.001839322,0.001886816]
u_values = [round(item * 10**6,2) for item in u_list]

n = len(u_values)
```

```python
Y = np.array(u_values).reshape(-1, 1)
X = np.column_stack([x1_coords, x2_coords])
```

```python
# Создаём объекты регрессии
ridge_model = Ridge(alpha=0.1)
lasso_model = Lasso(alpha=0.1)
```

```python
# Генерируем полиномиальные признаки
degree = 2 # Степень полинома
poly = PolynomialFeatures(degree)
```

```python
# Строим модели
ridge_pipeline = Pipeline([('poly',poly),('ridge', ridge_model)])
lasso_pipeline = Pipeline([('poly',poly),('lasso', lasso_model)])

ridge_pipeline.fit(X, Y)
lasso_pipeline.fit(X, Y)
```

```python
print("Ридж-регрессия")

Res_R = ridge_pipeline.predict(X)
for i in range(len(Res_R)):
    Res_R[i] = round(Res_R[i][0],2)

# вычисляем rmse
e_hat = []
for i in range(0, n):
    e_hat.append((u_values[i] - Res_R[i][0])**2)

sum_e = sum(e_hat)
rmse = round(sqrt((sum_e/n)),2)
print('rmse ', rmse)
```

```python
print("Лассо-регрессия")

Res_L = lasso_pipeline.predict(X)
for i in range(len(Res_L)):
    Res_L[i] = round(Res_L[i],2)

# вычисляем rmse
e_hat = []
for i in range(0, n):
    e_hat.append((u_values[i] - Res_L[i])**2)

sum_e = sum(e_hat)
rmse = round(sqrt((sum_e/n)),2)
print('rmse ', rmse)
```
